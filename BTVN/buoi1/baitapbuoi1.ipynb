II, Toán tử trong Python

1, Toán tử số học trong Python
- +, -,*,/,%
- ** : phép lấy số mũ ( 2**3 =2^3=8 )
- // : Phép toán chia số hạng trong đó kết quả là thương số trong đó các chữ số sau dấu thập phân đc loại bỏ. Nhưng nếu một trong các toán hạng là âm, kết quả sẽ đc làm tròn, đc làm tròn từ 0 ( 9//2=4, -9//4=-3, -11.0//3=-4.0 )

2, Toán tử quan hệ trong python
- ==, != (tương tự) <>, >, <, >=, <=


3, Toán tử gán trong Python
- =, +=, -=, *=, /=, &= (giống c++)
- **= Phép gán lũy thừa ( c** = a  là  c = c**a )
- //= Phép gán chia floor ( c//a la c = c//a )

4, Toán tử logic trong Python
- and giống &&  (a>9 and b<19) is true
- or giống || ( a>9 || b>19 ) is true
- not giống !  Not(a>9) là false

5, Toán tử menbership trong Python
- in : trả về true nếu tìm thấy biến a trong tập cấc giá trị sau in, ngược lại trả về false
- not in : trả về false nếu tìm thấy biến a trong tập các giá trị sau in, ngược lại trả về false

6, Toán tử nhận dạng (identify) trong pythonn
- is: trả về true nếu 2 biến cùng trỏ về 1 đối tượg giống nhau
- is not: trả về false nếu 2 biến ko cùng c trỏ về 1 đối tượng, ngược lại là true

7, Toán tử thay tác bit trong python
- &: sao chép một bit tới kết quả nếu bit này tồn tại trong cả hai toán hạng
- | : sao chép một bit tới kết quả nếu bit này tồn tải ở bất kỳ toán hạng nào
- ^ : sao chép bit nếu nó đc set (chỉ bit 1) chỉ trong một toán hạng
- ~ : Đây là toán tử một ngôi, đc sử dụng để đảo ngược bit
- << : toán tử dịch trái nhị phân. Giá trị của toán hạng trái được dịch chuyển sang trái một số lượng bit đã đc xác định bởi toán hạng phải.
- >> :  toán tử dịch phải nhị phân. Giá trị của toán hạng trái đc dịch chuyển sang phải một số lượng bit đã đc xác định bởi toán hạng phải.
